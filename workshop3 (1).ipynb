```python
line_data=df.groupby('Month')['ArrDelay'].mean().reset_index()
```

```python
line_data
```

```python
fig=go.Figure(data=go.Scatter(x=line_data['Month'],y=line_data['ArrDelay'],mode='lines',marker=dict(color='green')))
fig.update_layout(title='Monthly average delay of flights',xaxis_title='Months',yaxis_title='Delay of flight arrival(in minutes)')
fig.show()
```
